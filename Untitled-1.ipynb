```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the simulation
num_vehicles = 15  # Number of vehicles
num_points = 100  # Number of points in each trajectory
roundabout_radius = 15  # Radius of the roundabout
noise_level = 0.5  # Noise level to make trajectories varied
speed_variation = 0.1  # Variation in vehicle speeds

# Initialize figure
plt.figure(figsize=(8, 8))
plt.title("Trajectories of Tracked Vehicles at a Roundabout")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")

# Generate random starting points and speeds for each vehicle
start_angles = np.linspace(0, 2 * np.pi, num_vehicles, endpoint=False)
speeds = np.random.normal(1.0, speed_variation, num_vehicles)

# Simulate and plot trajectories
for i in range(num_vehicles):
    # Generate points along a circular path with added random noise
    angle = start_angles[i] + speeds[i] * np.linspace(0, 2 * np.pi, num_points)
    x = (roundabout_radius + noise_level * np.random.randn(num_points)) * np.cos(angle)
    y = (roundabout_radius + noise_level * np.random.randn(num_points)) * np.sin(angle)
    
    # Plot each trajectory
    plt.plot(x, y, marker="o", markersize=3, linestyle="-", alpha=0.7, label=f"Vehicle {i+1}")

# Display plot
plt.legend(loc="upper left", fontsize="small", ncol=2, bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the simulation
num_vehicles = 15  # Number of vehicles
num_points = 100  # Number of points in each trajectory
roundabout_radius = 5  # Radius of the roundabout
noise_level = 0.5  # Increased noise level to make trajectories more varied
outer_radius = 15  # Radius for entry/exit points
lane_offset = 0.01  # Small offset between lanes in each direction

# Entry/exit angles for paired lines in each direction (north, east, south, west)
entry_exit_angles = [
    0, lane_offset,               # Right (east)
    np.pi / 2, np.pi / 2 + lane_offset,        # Top (north)
    np.pi, np.pi + lane_offset,   # Left (west)
    3 * np.pi / 2, 3 * np.pi / 2 + lane_offset  # Bottom (south)
]

# Initialize figure
plt.figure(figsize=(8, 8))
plt.title("Trajectories of Tracked Vehicles at a Roundabout")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")

# Draw the central roundabout circle
roundabout_circle = plt.Circle((0, 0), roundabout_radius, color='gray', fill=False, linestyle='--', linewidth=1)
plt.gca().add_patch(roundabout_circle)

# Draw entry and exit lines connecting to the roundabout
for angle in entry_exit_angles:
    x_entry_exit = [outer_radius * np.cos(angle), roundabout_radius * np.cos(angle)]
    y_entry_exit = [outer_radius * np.sin(angle), roundabout_radius * np.sin(angle)]
    plt.plot(x_entry_exit, y_entry_exit, color='black', linestyle='-', linewidth=1)

# Simulate and plot only dots for each vehicle's trajectory
for i in range(num_vehicles):
    # Randomly assign entry and exit points (ensuring they are not the same direction)
    entry_angle = np.random.choice(entry_exit_angles)
    exit_angle = np.random.choice([angle for angle in entry_exit_angles if angle != entry_angle])
    
    # Generate points along the entry path to the roundabout
    entry_x = np.linspace(outer_radius * np.cos(entry_angle), roundabout_radius * np.cos(entry_angle), num_points // 4)
    entry_y = np.linspace(outer_radius * np.sin(entry_angle), roundabout_radius * np.sin(entry_angle), num_points // 4)
    
    # Generate circular path inside the roundabout
    circular_angle = np.linspace(entry_angle, exit_angle, num_points // 2)
    roundabout_x = roundabout_radius * np.cos(circular_angle) + noise_level * np.random.randn(num_points // 2)
    roundabout_y = roundabout_radius * np.sin(circular_angle) + noise_level * np.random.randn(num_points // 2)
    
    # Generate points along the exit path from the roundabout
    exit_x = np.linspace(roundabout_radius * np.cos(exit_angle), outer_radius * np.cos(exit_angle), num_points // 4)
    exit_y = np.linspace(roundabout_radius * np.sin(exit_angle), outer_radius * np.sin(exit_angle), num_points // 4)
    
    # Concatenate all segments to create the full trajectory
    x = np.concatenate((entry_x, roundabout_x, exit_x))
    y = np.concatenate((entry_y, roundabout_y, exit_y))
    
    # Plot each trajectory as larger dots without lines
    plt.plot(x, y, marker="o", markersize=6, alpha=0.7, linestyle="None")

# Display plot with appropriate limits
plt.xlim(-outer_radius - 2, outer_radius + 2)
plt.ylim(-outer_radius - 2, outer_radius + 2)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

# Parameters for the simulation
num_vehicles = 50  # Increase the number of vehicles
num_points = 150  # More points for smoother trajectories
roundabout_radius = 5  # Radius of the roundabout
noise_level = 0.3  # Increased noise level for varied trajectories
outer_radius = 15  # Radius for entry/exit points
lane_offset = 0.0021  # Slightly larger offset between lanes

# Entry/exit angles for paired lines in each direction (north, east, south, west)
entry_exit_angles = [
    0, lane_offset,               # Right (east)
    np.pi / 2, np.pi / 2 + lane_offset,        # Top (north)
    np.pi, np.pi + lane_offset,   # Left (west)
    3 * np.pi / 2, 3 * np.pi / 2 + lane_offset  # Bottom (south)
]

# Initialize figure
plt.figure(figsize=(8, 8))
plt.title("Trajectories of Tracked Vehicles at a Roundabout")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")

# Draw the central roundabout circle
roundabout_circle = plt.Circle((0, 0), roundabout_radius, color='gray', fill=False, linestyle='--', linewidth=1)
plt.gca().add_patch(roundabout_circle)

# Draw entry and exit lines connecting to the roundabout
for angle in entry_exit_angles:
    x_entry_exit = [outer_radius * np.cos(angle), roundabout_radius * np.cos(angle)]
    y_entry_exit = [outer_radius * np.sin(angle), roundabout_radius * np.sin(angle)]
    plt.plot(x_entry_exit, y_entry_exit, color='black', linestyle='-', linewidth=1)

# Simulate and plot trajectories for each vehicle
for i in range(num_vehicles):
    # Randomly assign entry and exit points, ensuring they are different directions
    entry_angle = np.random.choice(entry_exit_angles)
    exit_angle = np.random.choice([angle for angle in entry_exit_angles if angle != entry_angle])
    
    # Slightly vary the number of points in each segment for more natural paths
    entry_segment_points = np.random.randint(num_points // 6, num_points // 4)
    roundabout_segment_points = np.random.randint(num_points // 3, num_points // 2)
    exit_segment_points = num_points - (entry_segment_points + roundabout_segment_points)
    
    # Generate points along the entry path to the roundabout
    entry_x = np.linspace(outer_radius * np.cos(entry_angle), roundabout_radius * np.cos(entry_angle), entry_segment_points)
    entry_y = np.linspace(outer_radius * np.sin(entry_angle), roundabout_radius * np.sin(entry_angle), entry_segment_points)
    
    # Generate circular path inside the roundabout with added noise
    circular_angle = np.linspace(entry_angle, exit_angle, roundabout_segment_points)
    roundabout_x = roundabout_radius * np.cos(circular_angle) + noise_level * np.random.randn(roundabout_segment_points)
    roundabout_y = roundabout_radius * np.sin(circular_angle) + noise_level * np.random.randn(roundabout_segment_points)
    
    # Generate points along the exit path from the roundabout
    exit_x = np.linspace(roundabout_radius * np.cos(exit_angle), outer_radius * np.cos(exit_angle), exit_segment_points)
    exit_y = np.linspace(roundabout_radius * np.sin(exit_angle), outer_radius * np.sin(exit_angle), exit_segment_points)
    
    # Concatenate all segments to create the full trajectory
    x = np.concatenate((entry_x, roundabout_x, exit_x))
    y = np.concatenate((entry_y, roundabout_y, exit_y))
    
    # Plot each trajectory as dots without lines
    plt.plot(x, y, marker="o", markersize=3, alpha=0.6, linestyle="None")

# Display plot with appropriate limits
plt.xlim(-outer_radius - 2, outer_radius + 2)
plt.ylim(-outer_radius - 2, outer_radius + 2)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
